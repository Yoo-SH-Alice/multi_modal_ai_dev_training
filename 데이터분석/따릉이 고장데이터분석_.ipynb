```python
# csv파일 경로를 저장
# 2024년 상반기 따릉이 고장신고 데이터
csvPath1 = '/content/drive/MyDrive/Multi Modal AI개발자 양성 과정/code/데이터분석수업실습/따릉이 데이터 분석/서울시 공공자전거 고장신고 내역_2401-2406.csv'
# 2024년 하반기 따릉이 고장신고 데이터
csvPath2 = '/content/drive/MyDrive/Multi Modal AI개발자 양성 과정/code/데이터분석수업실습/따릉이 데이터 분석/서울시 공공자전거 고장신고 내역_2407-2412.csv'
```

```python
# prompt: 위에 지정한 경로 csvPath1, csvPath2를 dataframe으로 만들어줘

df1 = pd.read_csv(csvPath1, encoding='cp949')
df2 = pd.read_csv(csvPath2, encoding='cp949')
```

```python
# prompt: df1과 df2를 concat으로 위아래로 합치고 인덱스는 리셋해서 적용

brokenBycicle = pd.concat([df1, df2], ignore_index=True)
brokenBycicle
```

```python
brokenBycicle.info()
```

```python
# brokenBycicle '등록일시'는 datetime형 변환하고 yyyy-mm-dd 포멧을 변경함
brokenBycicle['등록일시'] = pd.to_datetime(brokenBycicle['등록일시']).dt.strftime('%Y-%m-%d')
brokenBycicle.head()
```

```python
# prompt: '등록일시'기준으로 groupby해서 '등록일자'와 '신고건수' 컬럼을 포함한 새로운 dataframe에 저장해줘

daily_reports = brokenBycicle.groupby('등록일시').size().reset_index(name='신고건수')
daily_reports = daily_reports.rename(columns={'등록일시': '등록일자'})
daily_reports
```

```python
# prompt: daily_reports로 x축은 '등록일자'로 y축은 '신고건수'로 산점도를 그려줘 y축은 라벨은 월별로만 표시해줘

daily_reports['등록일자'] = pd.to_datetime(daily_reports['등록일자'])

plt.figure(figsize=(15, 7))
plt.scatter(daily_reports['등록일자'], daily_reports['신고건수'])

plt.title('일별 따릉이 고장 신고 건수')
plt.xlabel('등록일자')
plt.ylabel('신고건수')

# x축 포맷을 월별로 설정
import matplotlib.dates as mdates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
